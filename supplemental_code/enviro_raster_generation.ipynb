Citharicthys Sordidus

```python
list_cit_sor=[]
names_cit_sor=[]

#add files to these lists
for root, dirs, files in os.walk(file_dir+"species_rasters_raw/Citharicthys_sordidus/"):
    for file in files:
        if file.endswith('.tif'):
            list_cit_sor.append(file_dir+"species_rasters_raw/Citharicthys_sordidus/"+file)
            name=file.replace(".tif","")
            names_cit_sor.append(name)
            names_cit_sor=natsorted(names_cit_sor,key=lambda y: y.lower())

for name in names_cit_sor:
    raster=rasterio.open(file_dir+"species_rasters_raw/Citharicthys_sordidus/"+name+".tif")
    #print(raster.meta) #inspect meta information
    #show(raster,title="%s"%name)

list_cit_sor_files=[]
for item in list_cit_sor:
    list_cit_sor_files.append(item)

names_cit_sor_files=[]
for item in names_cit_sor:
    names_cit_sor_files.append(item)


with open(file_dir+'species_rasters_raw/Citharicthys_sordidus/variable_list.txt', 'w') as f:
    for item in names_cit_sor_files:
        f.write("%s\n" % item)     


```
